### Time Series Analysis : MA Model

#### MA(q) - Moving average model of order q

Moving Average model is one where $Y_t$ depends only on the lagged forecast errors

$$Y_{t} = \mu + \varepsilon_{t} + \theta_{1}\varepsilon_{t-1} +... + \theta_{q}\varepsilon_{t-1} $$

where $\mu$ is the mean of the series, the $\theta_1, ..., \theta_q$ are the parameters of the model and the $\varepsilon_{t}, \varepsilon_{t−1},..., \varepsilon_{t−q}$ are white noise error terms. 

**Contrary to the AR model, the finite MA model is always stationary.**

#### Example 1. Simulate MA(1) process

```python
# import the module for simulating data
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline  
import warnings
warnings.filterwarnings("ignore")
sns.set_style("whitegrid")
from statsmodels.tsa.arima_process import ArmaProcess
from statsmodels.graphics import tsaplots
```

##### ACF and PACF plots of MA(1) process

```python
ar = np.array([1])
ma1 = np.array([1, 0.5]) 
MA_object1 = ArmaProcess(ar, ma1)
simulated_data_1 = MA_object1.generate_sample(nsample=500)
ma2 = np.array([1, 1.5])
MA_object2 = ArmaProcess(ar, ma2)
simulated_data_2 = MA_object2.generate_sample(nsample=500)
title_1 = r'$Y_{t} = \mu + \varepsilon_{t}+ 0.5\varepsilon_{t-1}$'
title_2 = r'$Y_{t} = \mu + \varepsilon_{t}+ 1.5\varepsilon_{t-1}$'

colors = ['darkseagreen', 'lightblue']
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(24, 14))
axes[0,0].plot(simulated_data_1, '.-', color=colors[0])
axes[0,0].set_title(title_1)
axes[0,1].plot(simulated_data_2, '.-', color=colors[1])
axes[0,1].set_title(title_2)
tsaplots.plot_acf(simulated_data_1, lags=50, 
                  ax=axes[1,0], title='ACF of '+ title_1)
tsaplots.plot_acf(simulated_data_2, lags=50, 
                  ax=axes[1,1], title='ACF of '+ title_2)

tsaplots.plot_pacf(simulated_data_1, lags=50, 
                  ax=axes[2,0], title='PACF of '+ title_1)
tsaplots.plot_pacf(simulated_data_2, lags=50, 
                  ax=axes[2,1], title='PACF of '+ title_1)

plt.tight_layout()
plt.show()
```

* Both ACF plots indicte stationary time series process.
* Both PACF plots indicte a order 1

#### Example 2. Simulate MA(2) process

```python
ar = np.array([1])
ma1 = np.array([1, 0.5, 0.8]) 
MA_object1 = ArmaProcess(ar, ma1)
simulated_data_1 = MA_object1.generate_sample(nsample=500)
ma2 = np.array([1, 0.9, 0.2])
MA_object2 = ArmaProcess(ar, ma2)
simulated_data_2 = MA_object2.generate_sample(nsample=500)
title_1 = r'$Y_{t} = \mu + \varepsilon_{t} + 0.5\varepsilon_{t-1} + 0.8\varepsilon_{t-2}$'
title_2 = r'$Y_{t} = \mu + \varepsilon_{t} + 0.9\varepsilon_{t-1} + 0.2\varepsilon_{t-2}$'

colors = ['plum', 'tan']
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(24, 14))
axes[0,0].plot(simulated_data_1, '.-', color=colors[0])
axes[0,0].set_title(title_1)

axes[0,1].plot(simulated_data_2, '.-', color=colors[1])
axes[0,1].set_title(title_2)

tsaplots.plot_acf(simulated_data_1, lags=50, 
                  ax=axes[1,0], title='ACF of '+ title_1)
tsaplots.plot_acf(simulated_data_2, lags=50, 
                  ax=axes[1,1], title='ACF of '+ title_2)

tsaplots.plot_pacf(simulated_data_1, lags=50, 
                  ax=axes[2,0], title='PACF of '+ title_1)
tsaplots.plot_pacf(simulated_data_2, lags=50, 
                  ax=axes[2,1], title='PACF of '+ title_1)

plt.tight_layout()
plt.show()
```
